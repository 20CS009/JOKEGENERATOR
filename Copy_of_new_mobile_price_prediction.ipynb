```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv('train.csv')
```

```python
dataset.head()
```

```python
dataset.describe()
```

```python
#internal memory vs price range
sns.pointplot(y='int_memory' , x='price_range', data=dataset)
```

```python
#ram vs price range
sns.jointplot(x='ram', y='price_range', data=dataset,color='red', kind='kde')
```

```python
#battery power vs price range
sns.boxplot(x='price_range', y='battery_power', data=dataset)
```

```python
#checking nan values
dataset.isna().sum()
```

```python
corr_mat=dataset.corr()
plt.figure(figsize=(16,16))

sns.heatmap(corr_mat,annot=True)
```

```python

X=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
```

```python
X[0]
```

```python
y[0]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier1=RandomForestClassifier(n_estimators=15,criterion='entropy')
classifier1.fit(x_train,y_train)
```

```python
y_pred1=classifier1.predict(x_test)
```

```python
y_pred1
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred1)
sns.heatmap(cm,annot=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier2=KNeighborsClassifier(n_neighbors=39,metric='minkowski')
classifier2.fit(x_train,y_train)
```

```python
y_pred2=classifier2.predict(x_test)
```

```python
y_pred2
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred2)
```
